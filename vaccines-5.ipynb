```python
import pandas as pd

dataset = pd.read_csv('/kaggle/input/vaccination/country_vaccination_stats.csv')

min_vacc = dataset.groupby("country")["daily_vaccinations"].transform("min")


dataset["daily_vaccinations"].fillna(min_vacc, inplace=True)
dataset["daily_vaccinations"].replace("nan", 0, inplace=True)
```

```python
median_vaccinations = dataset.groupby("country")["daily_vaccinations"].median()

top_3 = median_vaccinations.sort_values(ascending=False).head(3)
print(top_3)
```
